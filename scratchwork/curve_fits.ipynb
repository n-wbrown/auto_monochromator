```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from scipy.optimize import curve_fit
output_notebook()
```

```python
np.random.seed(3)
```

```python
def math_function(data,a,b):
    data = a * np.sin(2 * np.pi * data[0]) + b * np.cos(2 * np.pi * data[1])
    return data

def math_functionc(data,a,b):
    return data[0] * a + data[1] * b
```

```python
x_range = np.arange(-2,2,.05)
y_range = np.arange(-2,2,.05)
x, y = np.meshgrid(x_range, y_range)
a = .1
b = .2
c = .3
d = .4
z_practice = math_function((x,y),a,b)
z_practice = z_practice + np.random.normal(loc=0, scale=.02, size=z_practice.shape)
print(x,y,z_practice,sep='\n~~~~~~~~~~~~~~\n')
print("***********")
print(x.flatten(),y.flatten(),z_practice.flatten(),sep='\n~~~~~~~~~~~~~~\n')
```

```python
p = figure(plot_width=600, plot_height=300, output_backend="webgl")
p.circle(x[0],z_practice[0],radius=.01)
show(p)
```

```python
%%timeit
popt, pcov = curve_fit(math_function,(x.flatten(),y.flatten()),z_practice.flatten())

# print(popt, "\n",pcov)
```
