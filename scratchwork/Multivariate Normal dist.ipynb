```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
'''
=================
3D wireframe plot
=================

A very basic demonstration of a wireframe plot.
'''

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(8, 6), dpi=60)
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
X, Y, Z = axes3d.get_test_data(0.05)

# Plot a basic wireframe.
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
plt.show()
```

```python
'''
=================
3D wireframe plot
=================

A very basic demonstration of a wireframe plot.
'''

import matplotlib.pyplot as plt


fig = plt.figure(figsize=(8, 6), dpi=60)
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
X, Y  = np.meshgrid(
    np.arange(-3,3,.1),
    np.arange(-3,3,.1),
    indexing = 'ij')
X_flat = X.flatten()
Y_flat = Y.flatten()
# Z = X + Y
points = np.array([X_flat,Y_flat]).T
Z_flat = st.multivariate_normal.pdf(x=points, mean=np.array([0,0]),cov=[[1,0],[0,1]])
Z = Z_flat.reshape((len(X),len(Y)))
Z[0,0] = .1
Z[len(X)-1,0] = -.1
# Plot a basic wireframe.
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```
